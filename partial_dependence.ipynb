```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets.california_housing import fetch_california_housing
```

```python
cal_housing = fetch_california_housing()
X, y = cal_housing.data, cal_housing.target
names = cal_housing.feature_names
est = MLPRegressor(activation="logistic")
est.fit(X, y)
```

# First Plot

```python
from sklearn_plot_api import plot_partial_dependence
features = [0, 1, 2, 3]
viz = plot_partial_dependence(est, X, features, feature_names=[names[i] for i in features])
```

## Adjust color of House age

```python
viz.artists_['HouseAge'].set_color('green')
viz.figure_
```

## Add y label to AveRooms

```python
viz.axes_['AveRooms'].set_ylabel('Parital dependence')
viz.figure_
```
